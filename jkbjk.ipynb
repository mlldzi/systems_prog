```python
import pandas as pd
from sklearn import *
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
import numpy as np
```

```python
df = pd.DataFrame({"x1":[[1,2,3] for i in range(4)], 
                  "x2":[1.2,1.332,11.2,1.],
                  "x3":['aaa', 'bb', 'a', 'c']})
```

```python
preprocessing.OneHotEncoder(categories=df.x3)
```

```python
encoder = preprocessing.OneHotEncoder()
sub = encoder.fit_transform(df.x3.to_numpy().reshape(-1,1))
df2 = pd.DataFrame(sub.toarray(), columns = encoder.get_feature_names_out())
```

```python
pd.concat([df,df2], axis=1)
```

```python
encoder = preprocessing.LabelEncoder()
sub = encoder.fit_transform(df.x3.to_numpy().reshape(-1,1))
df['labeled_x3'] = sub
df
```
